# Analyse how the income of a patient affects the no.of visits to the hospital

```python
#relation between income and visits
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')
```

# Visualize the percentage of people getting government health insurance due to low income,due to old age and also the Percentage of people having private health insurance

```python
# % of people getting govt health insurance due to low income
label=['yes','no']
y=df[df['freepoor']=='yes']
n=df[df['freepoor']=='no']
x=[y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title(" % of people getting govt health insurance due to low income")
plt.show()



# % of people getting govt insurance due to old age,disability or veteran status
y=df[df['freerepat']=='yes']
n=df[df['freerepat']=='no']
x=[y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title(" % of people getting govt insurance due to old age,disability or veteran status ")
plt.show()

# % of People having private insurance
z=df[df['private']=='yes']
m=df[df['private']=='no']
x=[z.shape[0],m.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title(" % of People having private insurance ")
plt.show()
```

# plot a horizontal bar chat to analyse the reduced days of activity due to illness based on gender

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
#creating the bar chart
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
#adding the aesthetics
plt.title('Bar chart')
plt.xlabel('reduced activity ')
plt.ylabel('gender')
# show the plot
plt.show()
```
